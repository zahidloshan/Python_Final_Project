```python
import  pandas  as pd
```

```python
df1 = pd.read_csv ('./TotalGrade/TotalGrade.csv')
df1 = df1.drop(columns="Unnamed: 0")
df1
```

```python
grade = df1[df1.columns[16:17]]
grade['count'] = grade.groupby('Grade')['Grade'].transform('count')
grade.drop_duplicates(subset ="Grade",inplace = True)
grade
```

```python
grade['Grade']
```

```python
grade['count']
```

```python
import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
x=grade['Grade']
y=grade['count']
plt.xlabel('Grade')
plt.ylabel('Students')
plt.title('Student Grade Analysis')
plt.bar(x,y,color='green')
plt.show()
```

```python
attendance = df1[df1.columns[14:15]]
attendance['count'] = attendance.groupby('Attendance Mark')['Attendance Mark'].transform('count')
attendance.drop_duplicates(subset ="Attendance Mark",inplace = True)
attendance
```

```python
attendance['Attendance Mark']*10
```

```python
attendance['count']
```

```python
x=attendance['Attendance Mark']*10
y=attendance['count']
plt.xlabel('Attendance % ')
plt.ylabel('Number of Students ')
plt.title('Student Attendance Graph')
plt.bar(x,y,color='red')
plt.show()
```
