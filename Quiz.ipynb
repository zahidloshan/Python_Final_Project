```python
import  pandas  as pd
```

```python
df = pd.read_excel('./Quizes/Quiz 1.xlsx', sheet_name='Sheet1')
```

```python
df
```

```python
column = df[df.columns[4:6]]
```

```python
column
```

```python
result = column['Name'] 
column[['ID','Extra']] = result.str.split("@", n=1, expand=True) 
column
```

```python
newcolumn = column[column.columns[1:3]]
```

```python
newcolumn
```

```python
df1 = pd.read_csv ('./TotalGrade/TotalGrade.csv')
df1
```

```python
for  x in df1.index:
    df1.loc [x, "quiz1"] = 0
    for  y in newcolumn.index:
       if df1.loc [x, "Student ID"]==newcolumn.loc [y, "ID"] :
          df1.loc [x, "quiz1"]=newcolumn.loc [y, "Total points"]
```

```python
df1
```
