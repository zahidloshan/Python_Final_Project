```python
import  pandas  as pd
```

```python
df = pd.read_csv ('./Attendance/Week_1_Lab.csv')
df
```

```python
df.drop_duplicates(subset ="Full Name",inplace = True)
df
```

```python
df = df.drop(df.index[0])
df
```

```python
df1 = pd.read_csv ('./TotalGrade/TotalGrade.csv')
df1
```

```python
for  x in df1.index:
    df1.loc [x, "labday1"] = 0
    for  y in df.index:
       if df1.loc [x, "Full Name"]==df.loc [y, "Full Name"] :
             df1.loc [x, "labday1"] = 1

```

```python
df1
```

```python
df1.to_csv('./TotalGrade/TotalGrade.csv')
```
